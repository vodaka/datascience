```python
import json
import matplotlib.pyplot as plt
from collections import Counter
path = 'python_for_data_analysis/pydata-book/ch02/usagov_bitly_data2012-03-16-1331923249.txt'
records = [json.loads(line) for line in open(path)]
time_zones = [rec['tz'] for rec in records if 'tz' in rec]
```

```python
# count timezone with Counter class
counts = Counter(time_zones)
print counts.most_common(10)
```

```python
# count timezone with pandas
import pandas as pd
from pandas import DataFrame
from pandas import Series
frame = DataFrame(records)
first_ten = frame['tz'][:10]
print first_ten
tz_counts = frame['tz'].value_counts()
print tz_counts[:10]
```

```python
# Then,we might want to make plot of this data with plotting library, matplotlib.You can do a bit of munging to fill in a
# substitue value for unknown and missing time zone data in the records. The fillna function can replace missing(NA) values
# and uknown values(empty string) can be replaced by boolean array index:
clean_tz = frame['tz'].fillna('Missing')
print clean_tz
clean_tz[clean_tz==''] = 'Uknown'
tz_counts = clean_tz.value_counts()
print tz_counts[:10]
```

```python
import matplotlib.pylab as plt
plt.show(tz_counts[:10].plot(kind='barh',rot=0))
```
